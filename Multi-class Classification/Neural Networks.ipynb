<h3>Impementing the feedforward computation for the neural network and returning the neural networks prediction</h3>

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def feedForwardPropagation(Theta1, Theta2, X):
    m = X.shape[0]
    ones = np.ones((m,1))
    X = np.hstack((ones,X))
    a1 = sigmoid(X @ theta1.T)
    a1 = np.hstack((ones, a1))
    h = a1 @ Theta2.T
    return h
```

```python
def predict(result):
    return np.argmax(h, axis=1)+1
```

```python
result = feedForwardPropagation(theta1, theta2, X)
```

```python
print('The train accuracy is: ')
np.mean(predict(result) == y.flatten()) * 100
```
