# Polynomial Curve Fitting

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

## Data generation
Fits sine function using polynomial regression. We also added noise which was sampled using guassian distribution
$$\textbf{y} = \sin(2\pi \textbf{x}) + \epsilon$$ 
<br>Mind that we do not need normalize univariate input in this case

```python
n = 100
domain = (-1, 1)

def sample(n, domain, add_noise=True):
    x = np.random.uniform(*domain, n) # sample x
    y = np.sin(2*np.pi*x) + np.random.normal(scale=0.1, size=n)*float(add_noise)
    return x, y

x, y = sample(n, domain)
```

```python
sns.scatterplot(x, y)
```

## Polynomial Regression (Ordinary Least Square)
We are going to fits the train set using polynomial regression using following equation.
$$y(x, \mathbf{w})=w_{0}+w_{1} x+w_{2} x^{2}+\ldots+w_{M} x^{M}=\sum_{j=0}^{M} w_{j} x^{j}$$
Polynomial Regression is linear model with respect to parameter w despite the order of the polynomial. Polynomial regression could use any combination of x and its higher orders.

```python
w = np.ones(3+1) # include bias
def poly(x, w):
    y = np.zeros(np.array(x).shape)
    y = y + w[0]
    for j in range(1,len(w)):
        y = y + w[j]*np.power(x, j)
    return y
```

# Error function
The goal of the learning algorithm is to minimize error function with respect to w. We use mean square error (as it is regression problem). Factor 1/2 is included for convience when we are taking derivative of the error function.
$$J(\mathbf{w})=\frac{1}{2} \sum_{n=1}^{N}\left\{y\left(x_{n}, \mathbf{w}\right)-t_{n}\right\}^{2} +\frac{\lambda}{2}\|\mathbf{w}\|^{2}$$

```python
def MSE(y, y_prd, l):
    return 1/2*sum((np.array(y)-np.array(y_prd))**2) + l/2*(np.linalg.norm(w[1:])**2)
```

Because E(w) is quadratic function, dE(w) is linear function thus E(w) only have one local minimum and we could initialize w = 0

## Learning Algorithm - Batch Gradient Decent
Gradient update rules where alpha is the learning rate
$$w_{j}:=w_{j}-\alpha \frac{\partial}{\partial w_{j}} J(w)$$
$$\frac{\partial J(w)}{\partial w_{j}}=\sum_{n=1}^{N}\left\{y\left(x_{n},  {w}\right)-t_{n}\right\}x_{n}^{j}$$


```python
def gradient(x, y, w):
    m = len(w) # m + 1
    dJ = np.ones(m) # dJ_0 = 1 
    for j in range(0, m):
        dJ[j] = 1/2*sum(np.multiply((poly(x, w)-y), np.power(x, j)))
    return dJ
```

```python
def batch_gradient_decent(w, x_train, y_train, xval, yval, alpha, l):
    y_prd = poly(x_train, w)
    train_error = []
    validation_error = []
    for xi, yi in list(zip(x_train, y_train)):
        dJ = gradient(x_train, y_train, w)
        w -= alpha*dJ
        train_error.append(MSE(y_train, poly(x_train, w), l))
        validation_error.append(MSE(yval, poly(xval, w), l))

    return train_error, validation_error

m = 5
w = np.random.uniform(-0.3, 0.3, m+1)
alpha = 0.0003
l = 0.01
n_train = 5000
n_val = 1000
x_train, y_train = sample(n_train, domain)
x_val, y_val = sample(n_val, domain)

err_t, err_v = batch_gradient_decent(w, x_train, y_train, x_val, y_val, alpha, l)
```

## Analyzing Performance
The performance of the trained model is analyzed by learning curve.

```python
# plot learning curve
df = pd.DataFrame({
    'epoch':np.linspace(1, n_train, n_train),
    'J_train': err_t,
    'J_val': err_v   
})
df = df.melt('epoch', var_name='cols',  value_name='vals')
sns.lineplot(x='epoch', y='vals', hue='cols', data = df)
```

## Model Selection
We can decides M(degree of polynomial) by comparing accuracy on validations sets

```python
lambda_vec = [0, 0.001, 0.003, 0.01, 0.03, 0.1,0.3, 1, 3, 10]
alpha_vec = [0.005, 0.001, 0.0005, 0.0001, 0.00005]

n_train = 100
n_val = 20

x_train, y_train = sample(n_train, domain)
x_val, y_val = sample(n_val, domain)

min_err = float('inf')
optimal = (-1, -1, -1)
w_opt = None
for m in range(3, 10):
    print('Testing on m: '+str(m))
    w = np.random.uniform(-0.3, 0.3, m+1)
    for l in lambda_vec:
        for a in alpha_vec:
            _, err_val = batch_gradient_decent(w, x_train, y_train, x_val, y_val, a, l)
            if np.average(err_val) < min_err:
                min_err = np.average(err_val)
                optimal = (m, l, a)
                w_opt = w
```

## Learning Algorithm - Normal Equation
With small datasets, we could also use normal equation. Normal equation provide us with optimal weights without computing gradient in iterative manner

$$
\theta=\left(X^{T} X\right)^{-1} \cdot\left(X^{T} y\right)
$$

```python
def normal_equation(x_train, y_train, degree):
    phi = np.zeros(shape=(len(x_train), degree))
    for i in range(degree):
        phi[:, i] = x_train ** i

    y = np.array(y_train).reshape(-1, 1)

    return np.linalg.pinv(phi) @ y

x_train, y_train = sample(n_train, domain)
x_val, y_val = sample(n_val, domain)

w_opt = normal_equation(x_train, y_train, 7)


y = poly(x, w_opt)
sns.scatterplot(x_train, y_train, color='red')
sns.lineplot(x, y)
    
```

## Bayesian Regression
We use same model as OLS
$$y(x, \mathbf{w})=w_{0}+w_{1} x+w_{2} x^{2}+\ldots+w_{M} x^{M}=\sum_{j=0}^{M} w_{j} x^{j}$$

We define reponse Yi from the model

$$
T_{i} | x_{i}, \mathbf{w}, \sigma^{2} \sim \operatorname{Normal}\left(y(x,w), \sigma^{2}\right), \quad i=1, \cdots, n
$$

where the likelihood of each ti given xi, w, and sig2 is given by

$$
p\left(t_{i} | x_{i}, w, \sigma^{2}\right)=\frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left(-\frac{\left(t_{i}-\left(y(x,w)\right)\right)^{2}}{2 \sigma^{2}}\right)
$$
